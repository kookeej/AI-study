# 1. Matplotlib의 기본 사용
```pyplot```모듈은 각각의 함수를 사용해서 그래프를 만들 수 있다. 뿐만 아니라 그래프 영역 생성, 선 표시, 레이블 표시 등의 작업도 수행할 수 있다.     
특히 ```plot()```은 선, 또는 마커 그래프 그리기에 사용되는 함수다.

````python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('y-label')
plt.show()
````

여기서 plot의 인자로 들어간 리스트는 y값들이라 가정하고 x값은 자동으로 만들어낸다.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 3])
````

두 개의 리스트가 들어가게 되면, 첫 번째 리스트는 x값, 두 번째 리스트는 y값이 된다.

그래프에 스타일을 줄 수도 있다. plot의 세 번째 인자에 선의 **색상**과 **형태**를 지정하여 입력할 수 있다.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
````

기본 포멧은 'b-'이다. 'b-'는 'blue'와 '-'를 의미하기 때문에 파란 선을 뜻한다. 여기서는 'ro'이기 때문에 빨간색 원형의 마커로 표시하겠다는 뜻이다.    
```axis()```는 축의 범위를 나타낼 수 있다. 인자로 들어가는 리스트는 ```[xmin, xmax, ymin, ymax]``` 형식으로 들어간다.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-')
plt.axis([0, 6, 0, 20])
plt.show()
````

선/마커 표시 형식은 matplotlib 홈페이지에 들어가면 자세히 살펴볼 수 있다.    
![](https://wikidocs.net/images/page/92083/set_marker_03.png)


한 그래프 영역에 여러 개의 그래프를 그릴 수 있다. 

````python
import numpy as np

t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

```np.arange(0., 5., 0.2)```는 0~5까지 0.2간격으로 균일하게 샘플링된 시간이다.

# 2. Matplotlib의 그래프 꾸미기
## 2.1. 축 레이블
```xlabel(), ylabel()``` 함수를 사용하면 x, y축에 대한 레이블을 표시할 수 있다.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
````

## 2.2. 축 범위
```axis()```를 사용하여 x, y축의 범위를 지정할 수 있다.    
axis()의 인자로 하나의 리스트를 넣어주는데 그 리스트의 형식은 [xmin, xmax, ymin, ymax]이다.

````python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.axis([0, 5, 0, 20])
plt.show()
````

## 2.3. 색상 지정하기
```plot()```의 마지막 인자에 'r', 'b', 또는 'g'를 하면 각각 빨간색, 파란색, 초록색의 선이 그려진다.
color 키워드 인자를 사용하면 더 다양한 색상의 이름을 지정할 수 있다.

## 2.4. 그래프 영역 채우기
그래프의 특정 영역을 채워 나타낼 수 있다. 그래프를 채우는 함수로는 대표적으로 아래 세 가지의 함수가 있다.
1. fill_between()
2. fill_betweenx()
3. fill()

> ### fill_between()
fill_between()의 인자로 x[1:3], y[1:3]을 차례대로 입력해본다.

````python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5)

plt.show()
````

> ### fill_betweenx()
y[2:4], x[2:4]를 차례대로 입력한다.

````python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)

plt.show()
````

다음은 두 개의 그래프 사이의 영역을 채워보자.


````python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
````

> ### fill()
fill()을 이용하여 임의의 영역을 채울 수 있다. fill()에 x, y값의 리스트를 입력해주면, 각 x, y값으로 정의되는 다각형 영역이 채워진다.

````python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color='lightgray', alpha=0.5)

plt.show()
````

# 3. 다양하게 그래프 나타내기
## 3.1. 여러 곡선 그리기
여러 개의 곡선을 하나의 그래프 영역에 나타낼 수 있다.


````python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'r--', a, a**2, 'bo', a, a**3, 'g--')
plt.show()
````

그래프에 다양한 스타일을 적용할 수 있다. ```marker```와 ```markersize```, ```linewidth```를 이용해서 스타일을 적용할 수 있다.

````python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.show()
````

## 3.2. 그리드 설정
```grid()```함수를 사용해여 그래프에 그리드를 표시할 수 있다.


````python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)

# grid 표시
plt.grid(True)

plt.show()
````

축을 지정하여 원하는 축의 그리드만 표시할 수도 있다.

````python
plt.grid(True, axis='y')

plt.show()
````

color, alpha, linestyle 파라미터를 사용해서 그리드 선을 디자인 할 수도 있다.

````python
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')

plt.show()
````

## 3.3. 눈금 표시하기
그래프에 눈금을 표시할 수도 있다. ```xticks(), yticks(), tick_params()``` 함수를 이용해서 눈금을 표시한다.


````python
import matplotlib.pyplot as plt
import numpy as np

plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
````

여기서 **labels** 파라미터를 사용하면 눈금 레이블을 명시적으로 지정할 수도 있다.

````python
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
````

또한, 눈금 스타일을 예쁘게 세팅할 수도 있다. 내용은 아래와 같다.    


* ```tick_params()``` 함수를 사용하면 **눈금의 스타일**을 다양하게 설정할 수 있습니다.

* ```axis```는 설정이 적용될 축을 지정합니다. {‘x’, ‘y’, ‘both’} 중 선택할 수 있습니다.

* ```direction```을 ‘in’, ‘out’으로 설정하면 눈금이 안/밖으로 표시됩니다. {‘in’, ‘out’, ‘inout’} 중 선택할 수 있습니다.

* ```length```는 눈금의 길이를 지정합니다.

* ```pad```는 눈금과 레이블과의 거리를 지정합니다.

* ```labelsize```는 레이블의 크기를 지정합니다.

* ```labelcolor```는 레이블의 색상을 지정합니다.

* ```top/bottom/left/right```를 True/False로 지정하면 눈금이 표시될 위치를 선택할 수 있습니다.

* ```width```는 눈금의 너비를 지정합니다.

* ```color```는 눈금의 색상을 지정합니다.

````python
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.tick_params(axis='x', direction='in', length=3, pad=6, labelsize=14, labelcolor='green', top=True)
plt.tick_params(axis='y', direction='inout', length=10, pad=15, labelsize=12, width=2, color='r')

plt.show()
````

## 3.4. 타이틀 설정
```title()```함수를 사용하여 그래프의 타이틀을 설정할 수 있다. 뿐만 아니라 **타이틀 표시 및 위치 조절**과 **폰트 및 스타일 설정**도 할 수 있다.

````python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(0, 2, 0.2)
plt.plot(a, 'g-')
plt.axis([0, 10, 0, 2])
plt.grid(True, axis='y', color='gray', alpha=0.5)

plt.title('Sample graph')

plt.show()
````

이제 이 타이틀의 위치를 세팅할 수 있다.

````python
plt.title('Sample graph', loc='right', pad=20)
plt.show()
````

```loc```은 타이틀의 위치를 나타낸다. 'left', 'right', 'center'로 설정할 수 있다.    
```pad```는 타이틀과 그래프와의 간격을 의미한다.    

타이틀의 폰트도 지정할 수 있다. ```fontdict```에 딕셔너리 형태로 폰트에 대한 설정을 할 수 있다.

````python
title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('Sample graph', fontdict=title_font, loc='right', pad=10)
plt.show()
````

* ```fontsize```: 정확한 사이즈 외에도 'smaller', 'x-larger' 등과 같은 상대적인 설정도 가능하다.
* ```fontweight```: normal, bold, heavy, light, ultrabold, ultralight

# 4. 여러가지 그래프
## 4.1. 막대 그래프
막대 그래프는 ```bar()``` 함수를 통해 표현할 수 있다.


````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
````

```bar()```x리스트와 values리스트를 넣어준다. x에는 [0, 1, 2]가 들어간다.     
그 다음 ```x.ticks()```에 x, years를 입력해주면 위 그래프와 같이 2017, 2018, 2019가 순서대로 표시된다.    

또한, 막대 그래프에 스타일을 줄 수도 있다.

* ```width```는 막대의 너비입니다. 디폴트 값은 0.8이며, 예제에서는 0.6으로 설정했습니다.

* ```align```은 틱 (tick)과 막대의 위치를 조절합니다. 디폴트 값은 ‘center’인데, ‘edge’로 설정하면 막대의 왼쪽 끝에 x_tick이 표시됩니다.

* ```color```는 막대의 색을 지정합니다.

* ```edgecolor```는 막대의 테두리 색을 지정합니다.

* ```linewidth```는 테두리의 두께를 지정합니다.

* ```tick_label```을 어레이 형태로 지정하면, 틱에 어레이의 문자열을 순서대로 나타낼 수 있습니다.

* ```log=True```로 설정하면, y축이 로그 스케일로 표시됩니다.

````python
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values, width=0.6, align='edge', color='springgreen', edgecolor='gray',
        linewidth=3, tick_label=years, log=True)
plt.xticks(x, years)
plt.show()
````

## 4.2. 수평 막대 그래프
수평 막대 그래프는 ```barh()``` 함수를 이용하여 그릴 수 있다.

````python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.barh(y, values, height=-0.6, align='edge', color='green', 
         edgecolor='gray', linewidth=3, tick_label=years, log=False)
plt.show()
````

## 4.3. 산점도
산점도(scatter plot)은 두 변수의 상관 관계를 직교 좌표계의 평면에 데이터를 점으로 표현하는 그래프다. ```scatter()```함수를 사용하여 산점도를 그릴 수 있다.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(999)         # 같은 난수를 재사용할 수 있다. 즉, 예측 가능한 난수 생성

N = 50
x = np.random.rand(N)       # 마커의 위치
y = np.random.rand(N)
colors = np.random.rand(N)      # 마커의 색
area = (30 * np.random.random(N))**2        # 마커의 면적

plt.scatter(x, y, s=area, c=colors, alpha=0.5)  # alpha는 마커 색의 투명도를 결정해준다.
````

따라서 위 코드에서 x는 0~1범위 내에서 50개의 무작위 값을 갖는다.


## 4.4. 3차원 산점도
```scatter()``` 함수를 이용해서 3차원 산점도를 그릴 수도 있다.    
3차원 그래프를 그리기 위해서 ```from mpl_toolkits.mplot3d import Axes3D```를 추가해준다.하지만 이 부분은 matplotlib 3.1.0 버전부터는 디폴트로 포함되기 때문에 따로 작성하지 않아도 된다.

````python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(999)
n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2
xs = np.array([(xmax-xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax-ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax-zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax-cmin) * np.random.random_sample() + cmin for i in range(n)])

plt.rcParams["figure.figsize"] = (6, 6)         # figure의 사이즈 지정
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')      # 3D axes를 만들기 위해 projection=3d를 해준다. 
ax.scatter(xs, ys, zs, c=color, marker='o', cmap='Greens')      # colormap은 녹색 계열
plt.show()
````

## 4.5. 히스토그램
히스토그램은 도수분포표를 그래프로 나타낸 것이다. 가로축은 계급, 세로축은 도수(횟수가 개수)를 나타낸다. 계급은 변수의 구간이며 서로 겹치지 않는다. ```hist()```함수를 이용해서 히스토그램을 그려볼 수 있다.

````python
import matplotlib.pyplot as plt

# 몸무게 값을 저장한 리스트
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
plt.hist(weight)
plt.show()
````

여러 개의 히스토그램을 그릴 수 있다.

````python
a = 2.0 * np.random.randn(10000) + 1.0      # 표준편차 2.0, 평균 1.0인 정규분포
b = np.random.standard_normal(10000)        # 표준정규분포
c = 20.0 * np.random.rand(5000) - 10.0      # -10.0에서 10.0 사이의 균일한 분포를 갖는 5000개의 임의의 값

# bins: 몇 개의 영역으로 쪼갤지
# density=True: 밀도함수가 되어서 막대의 아래 면적이 1이 된다.
# alpha: 투명도
# histtype='step': 막대 내부가 비어있음
# histtype='stepfileed': 막대 내부가 채워져 있음
plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.show()
````

## 4.6. 에러바
에러바는 오차막대라고도 한다. 데이터의 편차를 표시하기 위한 그래프 형태이다. ```errorbar()``` 함수를 이용하여 에러바 그래프를 그릴 수 있다.


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [2.3, 3.1, 1.7, 2.5]          # 데이터의 편차를 나타내는 리스트

plt.errorbar(x, y, yerr=yerr)
plt.show()
````

yerr의 각 값들은 데이터 포인트의 위-아래 대칭인 오차로 표시가 된다.    

비대칭 편차를 나타낼 수 있다.

````python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(2.3, 3.1, 1.7, 2.5), (1.1, 2.5, 0.9, 3.9)]     # 첫번째 튜플은 아래 방향 편차, 두 번째 튜플은 위 방향 편차를 나타낸다.

plt.errorbar(x, y, yerr=yerr)
plt.show()
````

```uplims, lolims```를 사용하여 상한/하한 기호를 선택적으로 표시할 수도 있다.


````python
x = np.arange(1, 5)
y = x**2
yerr = np.linspace(0.1, 0.4, 4)
print(yerr)

plt.errorbar(x, y+4, yerr=yerr)
plt.errorbar(x, y+2, yerr=yerr, uplims=True, lolims=True)

upperlimits = [True, False, True, False]
lowerlimits = [False, False, True, True]
plt.errorbar(x, y, yerr=yerr, uplims=upperlimits, lolims=lowerlimits)
plt.show()
````

True면 표시, False면 표시하지 않겠다는 의미다.

## 4.7. 파이 차트
파이 차트는 범주별 구성 비율을 원형으로 표현한 그래프다. ```pie()```함수를 이용하여 파이 차트를 그릴 수 있다.

````python
import matplotlib.pyplot as plt
ratio = [34, 32, 16, 18]                    # 각 영역의 비율
labels = ['Apple', 'Banana', 'Melon', 'Grapes']     # 각 영역의 레이블

# autopct는 부채꼴 안에 표시될 숫자의 형식이다. 소수점 한자리까지만 표시하도록 한다.
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
````

```startangle```은 부채꼴이 그려지는 시작 각도를 세팅한다. 디폴트값은 0으로 설정되어 있고 이는 양의 방향 x축이다. ```counterclock```은 시계방향 순서로 부채꼴 영역을 표시할지를 결정한다.

````python
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
````

다음은 중심에서 벗어나는 정도를 설정할 수 있다. ```explode```는 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정하는 파라미터다.

````python
explode = [0, 0.10, 0, 0.10]
# Banana와 Grapes 영역에 대해서 반지름의 10%만큼 벗어나도록 설정했다.

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
````

파이 차트가 그림자를 나타낼 수 있도록 꾸밀 수도 있다. ```shadow=True```로 설정해주면 된다.

````python
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.show()
````

색상을 예쁘게 지정할 수도 있다. ```colors```를 사용하면 각 영역의 색상을 자유롭게 지정할 수 있다. ```Hex code```를 이용하여 지정할 수도 있다.

````python
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,
        explode=explode, shadow=True, colors=colors)
plt.show()
````

마지막으로 부채꼴 스타일을 지정할 수도 있다. ```wedgeprops```를 사용한다. ```wedgeprops``` 딕셔너리의 width, edgecolor, linewidth 키를 이용하여 각각 부채꼴 영역의 너비(반지름에 대한 비율), 테두리의 색상, 테두리의 선의 너비를 설정할 수 있다.

````python
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops = {'width': 0.7,
            'edgecolor': 'w', 
            'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,
        colors=colors, wedgeprops=wedgeprops)
plt.show()
````
